```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df=pd.read_csv("DecisionTreesClassificationDataSet.csv", sep=",")
df.head()
```

```python
df.isnull().sum()
```

```python
yesNo={"N":0,"Y":1}
egitim={"BS":0,"MS":1,"PhD":2}
df["SuanCalisiyor?"]=df["SuanCalisiyor?"].map(yesNo)
df["Top10 Universite?"]=df["Top10 Universite?"].map(yesNo)
df["StajBizdeYaptimi?"]=df["StajBizdeYaptimi?"].map(yesNo)
df["IseAlindi"]=df["IseAlindi"].map(yesNo)
df["Egitim Seviyesi"]=df["Egitim Seviyesi"].map(egitim)
```

```python
df.head()
```

```python
y=df["IseAlindi"]
x=df.drop(["IseAlindi"], axis=1)
clf=tree.DecisionTreeClassifier()
clf=clf.fit(x,y)
```

```python
print(clf.predict([[0.7,1,1,0,1,0]]))
```

```python
print(clf.predict([[0.7,1,1,0,0,0]]))
```
